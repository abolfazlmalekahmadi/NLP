```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#
 <h4 dir="rtl">بخش دوم 
 </h4>

<div dir="rtl">
در این قسمت ابتدا با ریکوست زدن به سايت مادر ، با استفاده از کتابخانه beautifuusoup قسمت های مورد نظر را با استفاده از تگ مورد نظر و کلاس خاص آن تگ به دست می اوریم و هچنین url  قسمت های مختلف برای دسترسی به آن قسمت و استخراج متن آن قسمت  دریافت و اصلاح میکنیم 
</div>

<div dir="rtl">

<b>پ.ن:</b> قسمت بلاگ های سایت(https://lastsecond.ir/blog) مورد توجه و هدف ما هست که  مطالب آن در حوزه گردشگری است</div>


```python
# Initialize an empty list to store the review URLs
review_urls = []

# Loop over the first 6 pages of the blog on lastsecond.ir
for page_number in range(6):
    # Construct the URL for the current page
    url = f"https://lastsecond.ir/blog?page={page_number}"
    
    # Send a request to the URL and get the response
    response = requests.get(url)
    
    # Parse the HTML content of the response using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the div element with class "post-list" that contains the articles
    article = soup.find("div", {"class": "post-list"})
    
    # Get the text of the article div and store it in the text variable
    text = article.get_text()
    
    # Loop over all the div elements with class "card post-list-item mb-3 card--hover" or "card post-list-item card--hover"
    for section in article.find_all("div", {"class": ["card post-list-item mb-3 card--hover",
                                                      "card post-list-item card--hover"]}):
        # Find the URL of the review in the h3 element and append it to the review_urls list
        review_urls.append(section.find("h3").find("a")["href"])

# Add the base URL "https://lastsecond.ir" to each review URL and append it to a new list called review_urls_new
review_urls_new=[]
for i in review_urls:
    review_urls_new.append("https://lastsecond.ir"+i)
    
# Return the review_urls_new list
review_urls_new




   

```

 # 
 <h4 dir="rtl">بخش سوم 
 </h4>
<div dir="rtl">
در این سلول و سلول های بعد به این صورت عمل میکنیم که  با ریکوست زدن به url های استخراج شد در قسمت قبل متن های مورد نظر را با استفاده از تگ خاص به کار رفته  برای قسمت های متن با استفاده از کلاس خاص ان تگ استخراج میکنیم </div>


```python
# Define a class called LastsecondReview
class LastsecondReview:
    # Define the class constructor
    # It takes a URL as input
    def __init__(self, url):
        # Set the input URL as an instance variable
        self.url = url
        # Send an HTTP request to the URL and get the web page content
        self.web_page = requests.get(self.url)
        # Create a BeautifulSoup object from the web page content
        self.soup = BeautifulSoup(self.web_page.content, "html.parser")
        # Call the get_description() and get_body() methods to extract the description and body of the review
        self.description = self.get_description()
        self.body = self.get_body()
        # Concatenate the description and body to get the full text of the review
        self.full_text = self.description + self.body
    
    # Define a method called get_description()
    def get_description(self):
        # Use BeautifulSoup to find the div with class "title" and extract its text
        aa = self.soup.find("div", {"class": "title"})
        return aa.get_text()
    
    # Define a method called get_body()
    def get_body(self):
        # Use BeautifulSoup to find the div with class "post-show__content__body typography" and extract its text
        return self.soup.find("div", {"class": "post-show__content__body typography"}).getText()

    
```

```python
review_list = []
for index, review_url in enumerate(review_urls_new):
    print(index+1, review_url)
    Lastsecond_review = LastsecondReview(review_url)
    review_list.append(Lastsecond_review)
```

```python
all_reviews = ""
for review in review_list:
    all_reviews += review.full_text + "\n"
print(len(all_reviews))
```

```python
reviews_file = open("review.txt", "w")
reviews_file.write(all_reviews)
reviews_file.close()
```

```python
review_df = pd.DataFrame(data={"url": [review.url for review in review_list], "text": [review.full_text for review in review_list]})

review_df.to_csv("review.csv", index=False)
```
