```python
from __future__ import unicode_literals
import pandas as pd
import random 
import numpy as np
import nltk
import pandas as pd
import codecs
import tqdm
from hazm import *
```

```python
normalizer = Normalizer()
```

```python
review_df = pd.read_csv(review_address)
review_df
```

```python
review_list = []
for index, row in review_df.iterrows():
  review_list.append(row["text"])
print(f"Number of reviews = {len(review_list)}")
```

```python
review_list_sent_tokenized = []
for review in review_list:
  review_list_sent_tokenized.append(sent_tokenize(review))
print(f"Number of sentences in the seconed review = {len(review_list_sent_tokenized[1])}")
print(f"First sentence of the seconed review:\n{review_list_sent_tokenized[1][0]}")
```

```python
stopwords = [normalizer.normalize(x.strip()) for x in codecs.open(stop_words_address,'r','utf-8').readlines()]
print(stopwords[:5])
```

```python
all_tokens_nonstop = [t for t in tqdm.tqdm(all_tokens) if t not in stopwords]
dataframe_nonstop = {}
dataframe_nonstop['all'] = FreqDist(eval(F"all_tokens_nonstop")).most_common(50)
freq_analysis_nonstop = pd.DataFrame(dataframe_nonstop)    
```

```python
freq_analysis_nonstop
```

```python
stemmer = Stemmer()
lemmatizer = Lemmatizer()

def get_lemma_set(tok, opt=1):
    if opt ==1:
        return stemmer.stem(tok)
    if opt ==2:
        return lemmatizer.lemmatize(tok)
    return
```

```python
all_tokens_nonstop_lemstem = [get_lemma_set(t, 1) for t in all_tokens_nonstop]
dataframe_nonstop_lemstem = {}
dataframe_nonstop_lemstem["all"] = FreqDist(eval(F"all_tokens_nonstop_lemstem")).most_common(50)
freq_analysis_nonstop_lemstem = pd.DataFrame(dataframe_nonstop_lemstem)   
freq_analysis_nonstop_lemstem
```

```python
tagger = POSTagger(model='postagger.model')
```

```python
import copy
dataframe_nonstop_lemstem_advjj = {}
selected_verb = []
selected_adj = []
selected_noun = []
selected_adv = []
for x,y in FreqDist(eval(F"all_tokens_nonstop_lemstem")).most_common(1000):
    pos = tagger.tag([x])[0][1]
    if pos in ['N']:
        selected_noun.append((x,y))
    if pos in ['AJ']:
        selected_adj.append((x,y))
    if pos in ['V']:
        selected_verb.append((x,y))
    if pos in ['ADV']:
        selected_adv.append((x,y))
dataframe_nonstop_lemstem_advjj["noun"]= copy.deepcopy(selected_noun[0:20])
dataframe_nonstop_lemstem_advjj["verb"]= copy.deepcopy(selected_verb[0:20])
dataframe_nonstop_lemstem_advjj["adj"]= copy.deepcopy(selected_adj[0:20])
dataframe_nonstop_lemstem_advjj["adv"]= copy.deepcopy(selected_adv[0:20])
dataframe_nonstop_lemstem_advjj = pd.DataFrame(dataframe_nonstop_lemstem_advjj)    
```
